```python
from script_utils import RESULT_BASE_DIR
import os
import json

MODEL = "custom-2"
```

```python
for domain in os.listdir(os.path.join(RESULT_BASE_DIR, MODEL)):
    domain_path = os.path.join(RESULT_BASE_DIR, MODEL, domain)
    if not os.path.isdir(domain_path):
        continue
    
    for task in os.listdir(domain_path):
        task_path = os.path.join(domain_path, task)
        if not os.path.isdir(task_path):
            continue

        with open(os.path.join(task_path, "result.txt"), "r") as f:
            result_txt_score = float(f.read().strip())
            
        with open(os.path.join(task_path, "summary", "results.json"), "r") as f:
            results_json = json.load(f)[0]
        
        if result_txt_score != results_json["score"]:
            results_json["score"] = result_txt_score

        if "error_reasons" in results_json:
            del results_json["error_reasons"]
        
        with open(os.path.join(task_path, "summary", "results.json"), "w") as f:
                json.dump([results_json], f, indent=4)
```
