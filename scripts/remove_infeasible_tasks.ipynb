```python
import script_utils as utils
import os
import json

MODEL = "custom-2"

tasks = utils.get_tasks(MODEL)
```

```python
infeasible_tasks = []
for domain in os.listdir(utils.EVALUATION_EXAMPLES):
    domain_path = os.path.join(utils.EVALUATION_EXAMPLES, domain)
    if not os.path.isdir(domain_path):
        continue
    print(f"Processing domain: {domain}")
    for task_file in os.listdir(domain_path):
        if not task_file.endswith(".json"):
            continue
        task_path = os.path.join(domain_path, task_file)
        with open(task_path, "r", encoding="utf-8") as f:
            task_data = json.load(f)
        evaluator_func = task_data.get("evaluator", {}).get("func") 
        if evaluator_func and evaluator_func == "infeasible":
            task_id = task_data["id"]
            infeasible_tasks.append(task_id)

print(f"Found {len(infeasible_tasks)} infeasible tasks.")
```

```python
# check whether model has such tasks 
model_infeasible_tasks = [task for task in tasks if task["task_id"] in infeasible_tasks]
print(f"Model {MODEL} has {len(model_infeasible_tasks)} infeasible tasks.")
```

```python
# remove them
for task in model_infeasible_tasks:
    utils.delete_run(MODEL, task["task_id"])
```
